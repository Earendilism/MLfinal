```python
class f1(sy.Function):
    @classmethod
    def eval(cls, x):
        return sy.N(0.3 * x**4 - 0.1 * x**3 - 2 * x**2 - 0.8 * x)


x = sy.Symbol("x")
lam_x = sy.lambdify(x, f1(x), modules=["numpy"])
x_vals = np.linspace(-3, stop=3, num=10000)
y_vals = lam_x(x_vals)
```

```python
class f3d(sp.Function):
    @classmethod
    def eval(cls, x, y):
        return (x**2 + y - 11) ** 2 + (x + y**2 - 7) ** 2


x, y = sp.symbols("x y")
r_min, r_max = -5.0, 5.0
# sample input range uniformly at 0.1 increments
xaxis = np.arange(r_min, r_max, 0.1)
yaxis = np.arange(r_min, r_max, 0.1)
# create a mesh from the axis
x_vals, y_vals = np.meshgrid(xaxis, yaxis)
lam_x = sp.lambdify((x, y), f3d(x, y), modules=["numpy"])
z_vals = lam_x(x_vals, y_vals)
```

## gradiant descent

```python
def gd(function, learning_rate, iter, x0=0.0):
    diff_y = function(x).diff(x)
    xlist = [x0]
    ylist = [function(x0)]
    while iter != 0:
        x0 = x0 - learning_rate * sy.N(diff_y.subs(x, x0))
        iter -= 1
        xlist.append(x0)
        ylist.append(function(x0))
    return xlist, ylist
```

```python
xoptgd, yoptgd = gd(function=f1, learning_rate=0.05, iter=100, x0=-2.8)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x_vals, y_vals)
ax.plot(xoptgd, yoptgd, "-*r")
plt.show()
```

```python
def momentum(function, learning_rate, iter, beta, x0=0.0):
    diff_y = function(x).diff(x)
    xlist = [x0]
    ylist = [function(x0)]
    m0 = 0
    while iter != 0:
        m0 = beta * m0 - learning_rate * sy.N(diff_y.subs(x, x0))
        x0 = x0 + m0
        iter -= 1
        xlist.append(x0)
        ylist.append(function(x0))
    return xlist, ylist
```

```python
xoptm, yoptm = momentum(function=f1, learning_rate=0.05, iter=100, beta=0.7, x0=-2.8)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x_vals, y_vals)
ax.plot(xoptm, yoptm, "-*r")
plt.show()
```

```python
def nesterovaccgradient(function, learning_rate, iter, beta, x0=0.0):
    diff_y = function(x).diff(x)
    xlist = [x0]
    ylist = [function(x0)]
    m0 = 0
    while iter != 0:
        m0 = beta * m0 - learning_rate * sy.N(diff_y.subs(x, x0 + beta * m0))
        x0 = x0 + m0
        iter -= 1
        xlist.append(x0)
        ylist.append(function(x0))
    return xlist, ylist
```

```python
xoptnag, yoptnag = nesterovaccgradient(
    function=f1, learning_rate=0.05, iter=100, beta=0.9, x0=-2.8
)
```

```python
def adagrad(function, learning_rate, iter, e, s0=0, x0=0.0):
    diff_y = function(x).diff(x)
    xlist = [x0]
    ylist = [function(x0)]
    while iter != 0:
        s0 = s0 + sy.N(diff_y.subs(x, x0)) * sy.N(diff_y.subs(x, x0))
        x0 = x0 - learning_rate * sy.N(diff_y.subs(x, x0)) / (s0 + e) ** 0.5
        iter -= 1
        xlist.append(x0)
        ylist.append(function(x0))
    return xlist, ylist
```

```python
xoptadag, yoptadag = adagrad(
    function=f1, learning_rate=0.05, iter=100, e=10.0**-10, s0=0, x0=-2.8
)
```

```python
def rmsprop(function, learning_rate, iter, s0=0, rho=0.0, x0=0.0):
    diff_y = function(x).diff(x)
    xlist = [x0]
    ylist = [function(x0)]
    while iter != 0:
        s0 = rho * s0 + (1 - rho) * sy.N(diff_y.subs(x, x0)) * sy.N(diff_y.subs(x, x0))
        x0 = x0 - learning_rate * sy.N(diff_y.subs(x, x0)) / (s0**0.5)
        iter -= 1
        xlist.append(x0)
        ylist.append(function(x0))
    return xlist, ylist
```

```python
xoptrmsprop, yoptrmsprop = rmsprop(
    function=f1, learning_rate=0.001, iter=100, s0=0, rho=0.5, x0=-2.8
)
```

```python
def adam(function, learning_rate, iter, s0=0, m0=0.0, beta1=0.0, beta2=0.0, x0=0.0):
    diff_y = function(x).diff(x)
    xlist = [x0]
    ylist = [function(x0)]
    while iter != 0:
        m0 = beta1 * m0 - (1 - beta1) * sy.N(diff_y.subs(x, x0))
        s0 = beta2 * s0 + (1 - beta2) * sy.N(diff_y.subs(x, x0)) * sy.N(
            diff_y.subs(x, x0)
        )
        m0 = m0 / (1 - beta1)
        s0 = s0 / (1 - beta2)
        x0 = x0 - learning_rate * m0 * sy.N(diff_y.subs(x, x0)) / ((s0 + 1) ** 0.5)
        iter -= 1
        xlist.append(x0)
        ylist.append(function(x0))
    return xlist, ylist
```

```python
xoptadam, yoptadam = adam(
    function=f1,
    learning_rate=0.05,
    iter=100,
    s0=0,
    m0=0.0,
    beta1=0.9,
    beta2=0.9,
    x0=-2.8,
)
```

```python
def gd3d(function, learning_rate, iter, x0=0.0, y0=0.0):
    diff_zx = function(x, y).diff(x)
    diff_zy = function(x, y).diff(y)
    xlist = [x0]
    ylist = [y0]
    zlist = [function(x0, y0)]
    while iter != 0:
        xp0 = x0 - learning_rate * diff_zx.subs(x, x0).subs(y, y0)
        y0 = y0 - learning_rate * diff_zy.subs(x, x0).subs(y, y0)
        x0 = xp0
        iter -= 1
        xlist.append(x0)
        ylist.append(y0)
        zlist.append(function(x0, y0))
    return xlist, ylist, zlist
```

```python
import sympy as sp

# Define the symbols
x, y = sp.symbols("x y")

# Define the function
f = (x**2 + y - 11) ** 2 + (x + y**2 - 7) ** 2

# Compute the gradient
f_grad = sp.Matrix([f.diff(var) for var in (x, y)])

# Define the gradient descent update


def gradient_descent_update(x_val, y_val, learning_rate):
    grad_val = np.array(f_grad.subs({x: x_val, y: y_val})).astype(float)
    res = np.array([0, 0]) - 0.05 * grad_val.T
    return res.flatten()[0], res.flatten()[1]


# Initialize variables
x_val, y_val = 0, 0
learning_rate = 0.01
iterations = 1000

# Gradient descent
for i in range(iterations):
    x_val, y_val = gradient_descent_update(x_val, y_val, learning_rate)

print(f"Minimum occurs at x = {x_val}, y = {y_val}")
```

```python
def gd(function, learning_rate: float = 0.01, iter: int = 100, init_val: dict = {}):
    f_grad = sp.Matrix([function.diff(var) for var in function.free_symbols])
    print(f_grad)
    grad_val = np.array(f_grad.subs(init_val)).astype(float)
    input_dict = {var: [] for var in function.free_symbols}
    while iter != 0:
        grad_val = np.array(f_grad.subs(init_val)).astype(float).T.flatten()
        print(grad_val)
        init_dict_val = list(init_val.values()) - learning_rate * grad_val
        inputdict = {
            var: input_dict[var] + [val]
            for var, val in zip(function.free_symbols, init_dict_val)
        }
        iter -= 1

    return input_dict
```
