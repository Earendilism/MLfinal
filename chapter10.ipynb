```python
import matplotlib.pyplot as plt
import numpy as np
import math


def exponential_decay_fn(epoch):
    return 0.01 * 0.5 * (epoch / 30)


def exponential_decay_fn_1(epoch):
    return 0.01 * 0.5 * (epoch // 30)


def exponential_decay_fn_2(epoch):
    return 0.01 * 0.5 * (np.ceil(epoch / 30))


x = np.linspace(0, 100, 1000)
y = exponential_decay_fn(x)
y_1 = exponential_decay_fn_1(x)
y_2 = exponential_decay_fn_2(x)
plt.plot(x, y)
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.legend(["exp_decay_fn", "exp_decay_fn_1", "exp_decay_fn_2"])
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist.load_data()
(x_train_full, y_train_full), (x_test, y_test) = fashion_mnist
x_train, y_train = x_train_full[:-5000], y_train_full[:-5000]
x_valid, y_valid = x_train_full[-5000:], y_train_full[-5000:]
x_train, x_valid, x_test = x_train / 255.0, x_valid / 255.0, x_test / 255.0
```

```python
fig, axes = plt.subplots(10, 10, gridspec_kw={"wspace": 0, "hspace": 0.5})
fig.set_figheight(15)
fig.set_figwidth(15)
dict_name = {
    0: "T-shirt/top",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle boot",
}
for i in range(10):
    for j in range(10):
        axes[i, j].imshow(x_train[10 * i + j], cmap="binary", vmax=1, vmin=0)
        axes[i, j].title.set_text(
            str(dict_name[y_train[10 * i + j]] + " " + str(y_train[10 * i + j]))
        )
        axes[i, j].axis("off")
        axes[i, j].set_xticklabels([])
        axes[i, j].set_yticklabels([])
plt.show()
fig, axes = plt.subplots(1, 1, gridspec_kw={"wspace": 0, "hspace": 0.5})
axes.imshow(x_train[7], cmap="binary", vmax=1, vmin=0)
axes.axis("off")
axes.title.set_text(str(dict_name[y_train[7]] + "is in class " + str(y_train[7])))
plt.show()
```

```python
tf.keras.backend.clear_session()
tf.random.set_seed(42)
model = tf.keras.Sequential()
model.add(tf.keras.layers.InputLayer(input_shape=[28, 28]))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(300, activation="relu"))
model.add(tf.keras.layers.Dense(200, activation="relu"))
model.add(tf.keras.layers.Dense(100, activation="relu"))
model.add(tf.keras.layers.Dense(10, activation="softmax"))
```

```python
model.compile(
    loss="sparse_categorical_crossentropy", optimizer="sgd", metrics=["accuracy"]
)
```

```python
history = model.fit(x_train, y_train, epochs=30, validation_data=(x_valid, y_valid))
```

```python
housing = fetch_california_housing()
X_train_full, X_test, y_train_full, y_test = train_test_split(
    housing.data, housing.target, random_state=42
)
X_train, X_valid, y_train, y_valid = train_test_split(
    X_train_full, y_train_full, random_state=42
)
```

```python
X_train_wide, X_train_deep = X_train[:, :5], X_train[:, 2:]
X_valid_wide, X_valid_deep = X_valid[:, :5], X_valid[:, 2:]
X_test_wide, X_test_deep = X_test[:, :5], X_test[:, 2:]
X_new_wide, X_new_deep = X_test_wide[:3], X_test_deep[:3]
```

```python
tf.keras.backend.clear_session()
```

```python
input_layer_deep = tf.keras.layers.Input(shape=[6], name="input_deep")
input_layer_wide = tf.keras.layers.Input(shape=[5], name="input_wide")
norm_deep = tf.keras.layers.Normalization()
norm_wide = tf.keras.layers.Normalization()
output_norm_deep = norm_deep(input_layer_deep)
output_norm_wide = norm_wide(input_layer_wide)
output_hidden_layer_deep_1 = tf.keras.layers.Dense(
    100, activation=tf.keras.activations.relu
)(output_norm_deep)
output_hidden_layer_deep_2 = tf.keras.layers.Dense(
    100, activation=tf.keras.activations.relu
)(output_hidden_layer_deep_1)
output_concat_layer = tf.keras.layers.Concatenate()(
    [output_norm_wide, output_hidden_layer_deep_2]
)
output_layer_aux = tf.keras.layers.Dense(1, name="aux")
output_layer_ord = tf.keras.layers.Dense(1, name="ord")
output_aux = output_layer_aux(output_hidden_layer_deep_2)
output_ord = output_layer_ord(output_concat_layer)
```

```python
model_m_input_m_output = tf.keras.Model(
    inputs=[input_layer_deep, input_layer_wide], outputs=[output_aux, output_ord]
)
```

```python
model_m_input_m_output.summary()
```

```python
tf.keras.utils.plot_model(
    model_m_input_m_output,
    show_shapes=True,
    show_layer_activations=True,
    expand_nested=True,
    show_trainable=True,
)
```

```python
model_m_input_m_output.compile(
    loss={
        "aux": tf.keras.losses.MeanSquaredError(),
        "ord": tf.keras.losses.MeanSquaredError(),
    },
    loss_weights={
        "aux": 0.9,
        "ord": 0.1,
    },
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
)
```

```python
norm_deep.adapt(X_train_deep)
norm_wide.adapt(X_train_wide)
```

```python
his_model_m_input_m_output = model_m_input_m_output.fit(
    x={"input_deep": X_train_deep, "input_wide": X_train_wide},
    y={"aux": y_train, "ord": y_train},
    epochs=10,
    validation_data=(
        {"input_deep": X_valid_deep, "input_wide": X_valid_wide},
        {"aux": y_valid, "ord": y_valid},
    ),
)
```
